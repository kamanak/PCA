Principle Component analyis from scratch

Step 1: Subtract the mean of each variable from the dataset so that the dataset should be centered on the origin and divideing with the standared deviation to get scaled value with mean 0 and varience 1.It will be helpful to calculate the covaricane

Step 2: Calculate the covariance matrix of the mean centered data

Step 3: Calculate the eigenvalues and eigen vectores for the calculated covarience matrix.The eigen vectors of the covarience matrix we get are Orthogonal to each other . Orthogonal means the vectors are mutually perpendicular to each other 

Step 4: Sort the eigenvalues in decending order along with their coresponding eigenvector. Each column in eigenvector matrx represent to principle component so arranging then n decending order of their eigenvalue will automatically arrange the principle component in decending order of their variability

Step 5: Select a subset from the rearranged eigenvalue i.e number of component= 2 . This means we selected only first two principle component 

Step 6: Finally transform the data by having a dot product between the transpose of the eigenvector subset and transpose of the mean centered data

```python
import numpy as np
def PCA(X,num_components):
    #Step-1

    X_scaled = (X - np.mean(X, axis=0))/np.std(X,axis=0)
    print('\n Value after subtracting mean : \n\n',X_scaled)

    #Step- 2 

    conv_mat = np.cov(X_scaled,rowvar = False)
    print('\n\n The calculated covariene matrix : \n\n',conv_mat)

    #Step_3
    eigen_values , eigen_vectors = np.linalg.eigh(conv_mat)
    print(f'\n\n The eigen_values are \n\n {eigen_values} \n\n  The eigen_vectors are \n\n {eigen_vectors}')

    #Step_4
    sorted_index = np.argsort(eigen_values)[::-1]
    print('\n\n The Sorted_index is \n\n',sorted_index)
    sorted_eigenvalue = eigen_values[sorted_index]
    print('\n\n Sorted eigenvalue is \n\n',sorted_eigenvalue)
    sorted_eigenvectors = eigen_vectors[:,sorted_index]
    print('\n\n Sorted eigenvectors is \n\n',sorted_eigenvectors)
  

    #Step_5
    eigenvector_subset = sorted_eigenvectors[:,0:num_components]
    print('\n\n Eigenvector subset according to number of component is \n\n ',eigenvector_subset)

    #Step 6
    X_reduced = np.dot(eigenvector_subset.transpose(),X_scaled.transpose()).transpose()
    print('X reduced is',X_reduced)
    return X_reduced


    
```

```python
import pandas as pd
 
#Get the IRIS dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
data = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
print('\n\n The given data is\n\n',data)
#prepare the data
x = data.iloc[:,0:4]
 
# #prepare the target
target = data.iloc[:,4]
 
# #Applying it to PCA function
mat_reduced = PCA(x , 2)
 


```

```python
# #Creating a Pandas DataFrame of reduced Dataset
principal_df = pd.DataFrame(mat_reduced , columns = ['PC1','PC2'])
principal_df
```

```python
 
# #Concat it with target variable to create a complete Dataset
principal_df = pd.concat([principal_df , pd.DataFrame(target)] , axis = 1)
principal_df
```
